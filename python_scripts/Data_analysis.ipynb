## T-test

```python
def t_test(ratings_a, ratings_p, threshold):
    
    t, p = stats.ttest_ind(ratings_a,ratings_p,equal_var=False)
    
    if p<threshold:
        print('We reject the null hypothesis')
        print("t = " + str(t))
        print("p = " + str(p))
    else:
        print('We accept the null hypothesis')
        print("t = " + str(t))
        print("p = " + str(p))
        
    print('First confidence interval :',st.t.interval(alpha=0.99, df=len(ratings_a)-1, loc=np.mean(ratings_a), scale=st.sem(ratings_a)) )
    print('Second confidence interval :',st.t.interval(alpha=0.99, df=len(ratings_p)-1, loc=np.mean(ratings_p), scale=st.sem(ratings_p)) )
    
    return 
```

```python
# loading data encoding from file

with open ('active_memories_list', 'rb') as fp:
    active_memories_list_M = pickle.load(fp)
```

```python
def calculate_ratio(memories_list):
    
    ratio_perceptual = []
    ratio_internal = []
    
    # iterate through users
    for i in range(len(memories_list)):
        
        nb_perceptual = 0
        nb_internal = 0
        nb_external = 0
        
        # Iterate through songs
        for j in range(NB_SONGS):
            
            nb_perceptual += int(memories_list[i][0][j])
            nb_internal += int(memories_list[i][1][j])

            nb_external += int(memories_list[i][2][j])
        
        # Calculate ratios 
        ratio_internal.append(nb_internal/(nb_internal+nb_external))
        ratio_perceptual.append(nb_perceptual/(nb_internal+nb_external))
        
    # returns ratio of perceptual and internal details
    return ratio_perceptual, ratio_internal

```

```python
scores_active_perceptual, scores_active_internal = calculate_ratio(active_memories_list_M)
```
